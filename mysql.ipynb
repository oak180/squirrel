## Users

```python
%%sql
SELECT persons.person_id, users.user_id, users.email, users.username, users.system_id, names.given_name, names.family_name, users.uuid
FROM ((users
INNER JOIN person AS persons ON users.person_id = persons.person_id)
INNER JOIN person_name AS names ON users.person_id = names.person_id)
;
```

## Patients

```python
%%sql
# SELECT *
SELECT patients.patient_id, names.given_name, names.family_name, people.uuid, patients.creator
FROM ((patient AS patients
INNER JOIN person_name AS names ON patients.patient_id = names.person_id)
INNER JOIN person AS people ON patients.patient_id = people.person_id)
ORDER BY patient_id ASC
;
```
